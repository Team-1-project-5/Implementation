### Dataset Background  

This dataset represents sales transactions recorded between **2011 and 2016** by a global retail company specializing in **bicycles, accessories, and clothing**. The data provides a comprehensive view of the company’s operations across multiple regions, including **Canada, Australia, the United States, France, Germany, and the United Kingdom**.  

Each transaction is described through several dimensions:  
- **Temporal information**: exact date, day, month, and year of the sale.  
- **Customer demographics**: age, gender, and age group segmentation.  
- **Geographical details**: country and state/region where the purchase was made.  
- **Product information**: product category, sub-category, and specific product name.  
- **Sales metrics**: order quantity, unit cost, unit price, total cost, revenue, and profit.  

The dataset enables analysis from multiple perspectives:  
- Identifying **customer segments** that drive the highest sales and profits.  
- Understanding **seasonality and temporal trends** in purchasing behavior.  
- Comparing **regional performance** across countries and states.  
- Evaluating **product-level profitability** to inform pricing and inventory strategies.  

Overall, this dataset provides a solid foundation for **sales performance analysis, customer behavior insights, and strategic decision-making** in the retail industry.

This code is used to check if the dataset contains any duplicate rows:

- **`data.duplicated()`** → Creates a Boolean Series where each row is marked as:  
  - `True` if it is a duplicate of a previous row.  
  - `False` if it is unique.  

- **`.sum()`** → Adds up all the `True` values, giving the total number of duplicate rows in the dataset.  

- **`print(f"Number of duplicate rows: {duplicates}")`** → Prints the result in a clear, formatted message using an f-string.  


````python
duplicates = data.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")
````

This code is used to **identify and display duplicate rows** in the dataset:

- **`data.duplicated(keep=False)`**  
  - Returns a Boolean Series where:  
    - `True` means the row is a duplicate (including the first occurrence).  
    - `False` means the row is unique.  
  - The argument `keep=False` ensures **all instances** of a duplicate are marked as `True`, not just the later ones.

- **`data[data.duplicated(keep=False)]`**  
  - Filters the dataset to include only the rows that are duplicates.  
  - The result is stored in the variable **`duplicates`**.  

- **`duplicates.shape[0]`**  
  - Returns the number of rows in the `duplicates` DataFrame, i.e., the total number of duplicate entries.  

- **`print(f"Found {duplicates.shape[0]} duplicate rows.")`**  
  - Prints the total count of duplicates in a clear message.  

- **`duplicates.head(10)`**  
  - Displays the first 10 duplicate rows for inspection.  

````python
duplicates = data[data.duplicated(keep=False)]

print(f"Found {duplicates.shape[0]} duplicate rows.")
duplicates.head(10)
````

This block of code removes duplicate rows from the dataset and then reports the new size:

- **`data.drop_duplicates()`**  
  - Removes all duplicate rows from the DataFrame.  
  - By default, it keeps the **first occurrence** of each duplicate and drops the rest.  
  - The result is reassigned to **`data`**, so the cleaned dataset overwrites the original one.

- **`data.shape[0]`** → Gives the updated number of rows after duplicates are removed.  
- **`data.shape[1]`** → Gives the total number of columns (unchanged).  

- **Print statement**  
  - Displays the new size of the dataset in a clear message using an f-string.  
  - This makes it easy to compare before and after duplicate removal.  

````python
data = data.drop_duplicates()

print(f"After removing duplicates, dataset has {data.shape[0]} rows and {data.shape[1]} columns.")
````

This block of code standardizes data types in the dataset to make sure each column is stored in the most appropriate format:


````python
data['Date'] = pd.to_datetime(data['Date'])
data['Customer_Age'] = data['Customer_Age'].astype(int)

cat_cols = ['Age_Group', 'Customer_Gender', 'Country', 'State',
            'Product_Category', 'Sub_Category', 'Product']

data[cat_cols] = data[cat_cols].astype('category')

float_cols = ['Unit_Cost', 'Unit_Price', 'Cost', 'Revenue', 'Profit']

data[float_cols] = data[float_cols].astype(float)
````

````python
data = data.drop(columns=['Date']) # delete the old date
````

````python
cols = ['Calculated_Date'] + [col for col in data.columns if col != 'Calculated_Date'] # adding the Calculated_Date to the beginging of the table
data = data[cols] 
````

````python
data.head()
````

````python
# This block of code creates a new column `Season` by mapping each date to its corresponding season:

def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Autumn'


data['Season'] = data['Calculated_Date'].dt.month.apply(get_season)
````

````python
data.head()
````

Creating Profit Margin

A new column called Profit_Margin is calculated using the formula:

# Profit Margin (%) = Profit / Revenue × 100 

Profit: The difference between Revenue and Cost.

Revenue: The total income generated from sales.

The result is multiplied by 100 to express it as a percentage.

````python
data['Profit_Margin'] = (data['Profit'] / data['Revenue']) * 100
data.head(100)
````

````python
# Age distribution
plt.hist(data['Customer_Age'], bins=20, color='skyblue', edgecolor='black')
plt.title("Customer Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

# Gender split
data['Customer_Gender'].value_counts().plot(kind='bar', color=['pink','blue'])
plt.title("Customer Gender Distribution")
plt.show()
````

````python
# Revenue by Gender
gender_revenue = data.groupby('Customer_Gender')['Revenue'].sum()
gender_revenue.plot(kind='bar', color=['pink','blue'], edgecolor='black')
plt.title("Revenue by Gender")
plt.show()

# Revenue by Age Group
age_revenue = data.groupby('Age_Group')['Revenue'].sum()
age_revenue.plot(kind='bar', color='purple', edgecolor='black')
plt.title("Revenue by Age Group")
plt.show()
````

````python
# Monthly revenue trend
monthly_revenue = data.groupby(data['Calculated_Date'].dt.to_period('M'))['Revenue'].sum()
monthly_revenue.index = monthly_revenue.index.to_timestamp()

plt.figure(figsize=(12,6))
plt.plot(monthly_revenue.index, monthly_revenue.values, marker='o', color='blue')
plt.title("Monthly Revenue Trend")
plt.xlabel("Date")
plt.ylabel("Revenue")
plt.grid(True, alpha=0.3)
plt.show()

````

````python
# Revenue by Country
country_revenue = data.groupby('Country')['Revenue'].sum().sort_values(ascending=False)
country_revenue.plot(kind='bar', figsize=(8,5), color='orange', edgecolor='black')
plt.title("Revenue by Country")
plt.ylabel("Revenue")
plt.show()
````

````python
# Top product categories
category_revenue = data.groupby('Product_Category')['Revenue'].sum().sort_values(ascending=False)
category_revenue.plot(kind='bar', color='green', edgecolor='black')
plt.title("Revenue by Product Category")
plt.ylabel("Revenue")
plt.show()

# Top 10 products by profit
top_products_profit = data.groupby('Product')['Profit'].sum().sort_values(ascending=False).head(10)
top_products_profit.plot(kind='barh', color='red', edgecolor='black')
plt.title("Top 10 Products by Profit")
plt.show()
````

````python
# Profit Margin Distribution
plt.hist(data['Profit_Margin'], bins=40, color='purple', edgecolor='black')
plt.title("Profit Margin Distribution")
plt.xlabel("Profit Margin (%)")
plt.ylabel("Frequency")
plt.show()
````
